```python
# Importing necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import files

# Step 1: Upload the CSV file
uploaded = files.upload()

# Load the CSV file into a DataFrame
df = pd.read_csv('StudentsPerformance.csv')

# Display the first few rows of the DataFrame
print(df.head())

# Step 2: Create visualizations

# 1. Distribution of Math Scores by Gender
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='math score', hue='gender', kde=True, palette='viridis')
plt.title('Distribution of Math Scores by Gender')
plt.xlabel('Math Score')
plt.ylabel('Frequency')
plt.show()

# 2. Boxplot of Reading Scores by Parental Education Level
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='parental level of education', y='reading score', palette='coolwarm')
plt.title('Reading Scores by Parental Level of Education')
plt.xlabel('Parental Level of Education')
plt.ylabel('Reading Score')
plt.xticks(rotation=45)
plt.show()

# 3. Countplot of Lunch Types
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='lunch', palette='Set2')
plt.title('Lunch Type Distribution')
plt.xlabel('Lunch Type')
plt.ylabel('Count')
plt.show()

# 4. Scatterplot of Writing vs. Reading Scores
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='reading score', y='writing score', hue='gender', style='lunch', palette='Set1')
plt.title('Writing vs Reading Scores')
plt.xlabel('Reading Score')
plt.ylabel('Writing Score')
plt.show()

# 5. Heatmap of Correlation Between Scores
plt.figure(figsize=(8, 6))
corr = df[['math score', 'reading score', 'writing score']].corr()
sns.heatmap(corr, annot=True, cmap='Blues')
plt.title('Correlation Between Scores')
plt.show()
```

```python
import os
import shutil

# Function to create the folder structure and files
def create_folder_structure_and_files():
    # Define the folder structure
    project_name = 'student_performance_project'
    folders = [
        os.path.join(project_name, 'data_raw'),
        os.path.join(project_name, 'data_clean'),
        os.path.join(project_name, 'results'),
        os.path.join(project_name, 'src')
    ]

    # Remove the project folder if it already exists
    if os.path.exists(project_name):
        shutil.rmtree(project_name)

    # Create folders
    for folder in folders:
        os.makedirs(folder, exist_ok=True)

    # Create raw data file and README
    with open(os.path.join(project_name, 'data_raw', 'StudentsPerformance.csv'), 'w') as f:
        f.write('gender,race/ethnicity,parental level of education,lunch,test preparation course,math score,reading score,writing score\n')
        f.write('female,group B,bachelor\'s degree,standard,none,72,74,74\n')
        f.write('female,group C,some college,standard,completed,69,88,86\n')
        f.write('female,group B,master\'s degree,standard,none,90,93,92\n')
        f.write('male,group A,associate\'s degree,free/reduced,none,47,54,44\n')
        f.write('male,group C,some college,standard,none,76,78,75\n')
        f.write('female,group B,associate\'s degree,standard,none,71,83,78\n')
        f.write('female,group B,some college,standard,completed,88,95,92\n')
        f.write('male,group B,some college,free/reduced,none,40,43,39\n')
        f.write('female,group B,some college,standard,none,72,78,76\n')
        f.write('female,group C,some college,standard,none,69,75,73\n')

    with open(os.path.join(project_name, 'data_raw', 'README.txt'), 'w') as f:
        f.write('This folder contains the raw data for student performance.\n')
        f.write('The main data file is StudentsPerformance.csv.\n')

    # Create cleaned data file
    with open(os.path.join(project_name, 'data_clean', 'clean_students_data.csv'), 'w') as f:
        f.write('gender,race/ethnicity,parental level of education,lunch,test preparation course,math score,reading score,writing score\n')

    # Create results file
    with open(os.path.join(project_name, 'results', 'analysis_results.txt'), 'w') as f:
        f.write('This file will contain the results of the data analysis.\n')

    # Create cleaning script
    with open(os.path.join(project_name, 'src', 'clean_data.py'), 'w') as f:
        f.write('''import pandas as pd

# Load the data
df = pd.read_csv('../data_raw/StudentsPerformance.csv')

# Data cleaning steps would go here
# For now, we will just save the cleaned data without modifications
df.to_csv('../data_clean/clean_students_data.csv', index=False)
print("Data cleaned and saved.")
''')

    # Create analysis script
    with open(os.path.join(project_name, 'src', 'analysis.py'), 'w') as f:
        f.write('''import pandas as pd

# Load the cleaned data
df = pd.read_csv('../data_clean/clean_students_data.csv')

# Analysis steps would go here
# For now, we will just print the first few rows
print(df.head())
''')

# Create the folder structure and files
create_folder_structure_and_files()
print("Folder structure and files created successfully!")
```
